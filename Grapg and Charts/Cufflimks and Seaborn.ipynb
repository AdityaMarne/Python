```python
sns.load_dataset('tips')
```

```python
df1 = sns.load_dataset('tips')
```

```python
sns.relplot(x = "total_bill", y="tip",data=df1,hue="smoker") #hue is trying to perform group by operation
```

```python
sns.relplot(x = "total_bill", y="tip",data=df1, style = "time",hue="smoker") 
```

```python
sns.relplot(x = "total_bill", y="tip",data=df1, size="size") 
```

```python
sns.relplot(x = "total_bill", y="tip",data=df1,kind = "line") 
```

```python
sns.relplot(x = "total_bill", y="tip", col = "day",data=df1) 
```

```python
sns.catplot(x = "total_bill", y="tip",data=df1) 
```

```python
sns.catplot(x = "day", y="total_bill",data=df1) 
```

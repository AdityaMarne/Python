# Box Plot
Box plot is used to find outliers. Outlier is a dataset which is not able to fit itself in between regular range/distribution of dataset.

```python
df.plot(kind = 'box',figsize=(20,30))
```

```python
df.plot(kind = 'box' ,figsize = (20,10),color = {'boxes' : 'g' , 'whiskers' :'r'} , vert = False )
```

# Area Graph

```python
df.plot(kind = 'area',figsize = (20,10), stacked = False)
```

# Scatter Plot
